```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

def create_professional_slide(text_input, output_name="slide.png"):
    # 1. Setup Canvas (Dark Navy Blue)
    width, height = 1920, 1080
    img = Image.new('RGB', (width, height), color=(15, 23, 42))
    draw = ImageDraw.Draw(img)

    # 2. Load a BIG Font
    # In Colab, this path usually works for a clean, bold look:
    font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf"
    font_size = 80
    font = ImageFont.truetype(font_path, font_size)

    # 3. Wrap the text so it doesn't stay in one long line
    # (width=40 means roughly 40 characters per line)
    wrapper = textwrap.TextWrapper(width=40)
    wrapped_text = wrapper.fill(text=text_input)

    # 4. Draw the text in the exact center
    # "mm" anchor stands for Middle-Middle
    draw.multiline_text((width/2, height/2), wrapped_text, font=font,
                        fill=(255, 255, 255), anchor="mm", align="center", spacing=20)

    img.save(output_name)
    print(f"Professional slide saved as {output_name}")

# Test it!
create_professional_slide("Welcome Srishti! This text is now large, centered, and wrapped properly for your training video.")
```

```python
def generate_srt(text, output_file="subtitles.srt"):
    # For now, we will create a "simple" SRT
    # In a full version, we use Kokoro's 'ps' (phonemes/segments) to get timing
    words = text.split()
    duration_per_word = 0.5  # Rough estimate for now

    with open(output_file, "w") as f:
        for i, word in enumerate(words):
            start = i * duration_per_word
            end = (i + 1) * duration_per_word

            # SRT Format: index, time range, text
            f.write(f"{i+1}\n")
            f.write(f"00:00:{start:05.2f} --> 00:00:{end:05.2f}\n".replace('.', ','))
            f.write(f"{word}\n\n")
    print(f"Subtitles saved to {output_file}")

generate_srt(socratic_text)
```

```python
import soundfile as sf
import math

def generate_perfect_srt(text, audio_file, output_file="subtitles.srt"):
    # 1. Get the actual length of the audio in seconds
    data, samplerate = sf.read(audio_file)
    total_duration = len(data) / samplerate

    words = text.split()
    # 2. Calculate the exact time per word based on total audio length
    duration_per_word = total_duration / len(words)

    with open(output_file, "w") as f:
        for i, word in enumerate(words):
            start = i * duration_per_word
            end = (i + 1) * duration_per_word

            # Formatting seconds into SRT time format (HH:MM:SS,mmm)
            def format_time(seconds):
                h = int(seconds // 3600)
                m = int((seconds % 3600) // 60)
                s = int(seconds % 60)
                ms = int((seconds % 1) * 1000)
                return f"{h:02}:{m:02}:{s:02},{ms:03}"

            f.write(f"{i+1}\n")
            f.write(f"{format_time(start)} --> {format_time(end)}\n")
            f.write(f"{word}\n\n")

    print(f"Perfectly synced subtitles saved! Total Duration: {total_duration:.2f}s")

# Run it using your previous synced audio
generate_perfect_srt(socratic_text, "synced_audio.wav")
```

```python
import soundfile as sf

def generate_karaoke_ass(text, audio_file, output_file="karaoke.ass"):
    # 1. Get audio duration
    data, samplerate = sf.read(audio_file)
    total_duration_ms = (len(data) / samplerate) * 100 # ASS uses centiseconds (1/100th of a second)

    words = text.split()
    duration_per_word = total_duration_ms / len(words)

    # 2. Build the ASS Header (Setting the Styles)
    header = """[Script Info]
ScriptType: v4.00+
PlayResX: 1920
PlayResY: 1080

[V4+ Styles]
Format: Name, Fontname, Fontsize, PrimaryColour, SecondaryColour, OutlineColour, BackColour, Bold, Italic, Underline, Strikeout, ScaleX, ScaleY, Spacing, Angle, BorderStyle, Outline, Shadow, Alignment, MarginL, MarginR, MarginV, Encoding
Style: Default,Arial,80,&H00FFFFFF,&H0000FFFF,&H00000000,&H00000000,1,0,0,0,100,100,0,0,1,2,0,2,10,10,100,1
"""
    # SecondaryColour (&H0000FFFF) is Yellow - this is what the word changes TO.

    # 3. Build the Karaoke Line
    events = "\n[Events]\nFormat: Layer, Start, End, Style, Name, MarginL, MarginR, MarginV, Effect, Text\n"

    # Start/End time for the whole sentence
    start_time = "0:00:00.00"
    end_time = f"0:00:{int(total_duration_ms/100):02d}.{int(total_duration_ms%100):02d}"

    karaoke_text = ""
    for word in words:
        # \k[duration] is the tag. Duration is in centiseconds.
        karaoke_text += f"{{\\k{int(duration_per_word)}}}{word} "

    line = f"Dialogue: 0,{start_time},{end_time},Default,,0,0,0,,{karaoke_text}"

    with open(output_file, "w") as f:
        f.write(header + events + line)

    print(f"Karaoke script '{output_file}' created!")

# Run it!
generate_karaoke_ass(socratic_text, "synced_audio.wav")
```
